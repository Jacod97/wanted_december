```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/질병관리청/성인흡연추이.csv')
data
```

```python
total_age = data.loc[:,'2012':'2022'][1:2].transpose()
total_age
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

total_age.astype(float).plot(kind='bar', figsize=[10,5]).get_legend().remove()
plt.xticks(rotation=0)
plt.title('성인흡연율 추이')
plt.yticks(range(0,31,5))
```

```python
adul_line = total_age.astype(float).plot(kind='line', figsize=[10,5]).get_legend().remove()
plt.xticks(rotation=0)
plt.title('성인흡연율 추이')
plt.yticks(range(0,31,5))
```

```python
student = pd.read_csv('data/질병관리청/청소년흡연추이.csv',index_col=0).dropna().reset_index(drop=True)
student
```

```python
student_age = student.iloc[:,2:][1:2].transpose()
student_age
```

```python
student_age.astype(float).plot(kind='bar', figsize=[10,5]).get_legend().remove()
plt.xticks(rotation=0)
plt.title('청소년 흡연 추이')
plt.yticks(range(0,21,5))
```

```python
student_line = student_age.astype(float).plot(kind='line', figsize=[10,5]).get_legend().remove()
plt.xticks(rotation=0)
plt.title('청소년 흡연 추이')
plt.yticks(range(0,21,5))
```

```python
smoke = pd.concat([data,student])
smoke
```

```python
smoke_total = smoke.iloc[:,3:12][1:5:3].transpose()
smoke_total.columns = ['성인', '청소년']
```

```python
smoke_total.astype(float).plot(kind='line', figsize=[10,5], color=['red', 'blue'], linewidth= 3)
plt.xticks(rotation=0)
plt.title('전체 흡연율 추이')
plt.yticks(range(0,31,5))
```
